##### Broadcasting

Broadcasting is TensorFlow’s way of performing element‑wise operations on tensors with different shapes as long as their shapes are compatible, meaning TensorFlow can “stretch” dimensions of size 1 so that two tensors can align for arithmetic operations. Broadcasting allows writing vectorized code without manually repeating or reshaping tensors.

**Broadcasting Rules**: Two dimensions are compatible when
- They are equal, or
- One of them is 1 
<br>

TensorFlow compares shapes from right to left (trailing dimensions first).
If all dimensions are compatible, TensorFlow automatically broadcasts the smaller tensor.

check more on broadcasting operations here https://numpy.org/doc/stable/user/basics.broadcasting.html


```python
# broadcasting with scalars
X = tf.constant([1., 2., 3.])
y = 2.0
print(f"{X} with shape {X.shape}")
print(f"scalar {y}")
print("X * y gives us: ", X * y)
print("y is broadcasted to shape internally (3,) → [2., 2., 2.]")
```

```python
# broadcasting a 1D tensor to a 2D tensor
X = tf.constant([
    [0., 0., 0.],
    [10., 10., 10.],
    [20., 20., 20.],
    [30., 30., 30.]
])  # shape (4, 3)
b = tf.constant([1., 2., 3.])  # shape (3,)
print("X + b =\n", X + b)
print("----------------")
print(
'''
X shape: (4, 3)
b original shape: (3,) → treated as (1, 3)
b is broadcast across rows → result shape (4, 3)
'''
)
```

```python
# broadcasting with inserted dimensions
a = tf.constant([0., 10., 20., 30.])  # shape (4,)
b = tf.constant([1., 2., 3.])         # shape (3,)
a_expanded = a[:, tf.newaxis] # Make a into shape (4, 1)
print("a_expanded shape:", a_expanded.shape)
print("b shape:", b.shape)

result = a_expanded + b
print("Result shape:", result.shape)
print(result)
print(
'''
a[:, None] → shape (4, 1)
b → shape (3,) → treated as (1, 3)
Result → (4, 3) outer addition
'''
)
```

```python
# broadcasting a 3D tensor with a 1D tensor
X = tf.random.uniform((4, 3, 2))  # shape (4, 3, 2)
b = tf.constant([1., 2.])         # shape (2,)
print("X shape:", X.shape)
print("b shape:", b.shape)
Y = X + b
print("Y shape:", Y.shape)
print(
'''
X last dimension = 2
b shape = (2,) → aligns with last dimension
b is broadcast across the first two axes → (4, 3, 2)
'''
)
```

```python
# broadcasting fails when trailing dimensions don’t match and neither is 1.
a = tf.constant([1., 2., 3.])   # shape (3,)
b = tf.constant([1., 2., 3., 4.])  # shape (4,)
try:
    print(a + b)
except Exception as e:
    print("Broadcasting error:", e)
```

##### Dimension Expension

```python
# adding dimension to tensor
np_X = np.arange(0,10, dtype=np.int32) 
X = tf.constant(np_X, shape=(1,10))
print(f"Original Tensor: {X}, with shape {X.shape} \n--------------")

# added new axis at the end
X = tf.constant(np_X, shape=(1,10))
X = X[..., tf.newaxis]
print(f"New tensor with shape {X.shape} \n------------")

# added new axis in the middle
X = tf.constant(np_X, shape=(5,2))
X = X[:,tf.newaxis,:]
print(f"New tensor with shape {X.shape} \n------------")

# 0 to expand first axis, 1 for second,...-1 to expand the final axis
np_X = np.arange(0,24, dtype=np.int32) 
X = tf.constant(np_X, shape=(4,3,2))
print(f"Original Tensor with shape {X.shape} \n--------------")
X = tf.expand_dims(X, axis=-1) 
print(f"New tensor with shape {X.shape}")
```

##### Matrix Multiplication

```python
A = tf.constant([[1, 1],
                 [2, 2],
                 [3, 3]])   # shape (3, 2)
B = tf.constant([[0, 0],
                 [1, 0],
                 [0, 1]])   # shape (3, 2)
print("A shape:", A.shape)
print("B shape:", B.shape)
print("\nElement-wise multiply:\n", A * B)
print("Both tensors have same shape (3, 2) → element‑wise operations are valid.")
```

```python
# but matrix multiplication matmul is different
try:
    tf.matmul(A, B)
except Exception as error:
    print(type(error).__name__, ":", error)
```

**A** is **3 × 2** and **B** is **3 × 2**

**Matrix multiplication rule:**
> (**m × n**) · (**p × q**) is valid only if **n = p**
**Here:**
- **A** inner dimension = **2**
- **B** inner dimension = **3**

Since **2 ≠ 3**, TensorFlow raises an error.


```python
# fix the above using transpose
B_T = tf.transpose(B)   # shape becomes (2, 3)
print(f"transposed from \n {B} \n to \n {B_T}" )
print("B_T shape:", B_T.shape)

result = tf.matmul(A, B_T)
print("\nA @ B_T =\n", result)
```

```python
# or we can reshape but result will be different (we get same shape but order of elements is different now)
B_reshaped = tf.reshape(B, (2, 3))
print(f"reshaped from \n {B} \n to \n {B_reshaped}")

result = tf.matmul(A, B_reshaped)
print("\nA @ B_T =\n", result)
```

Transpose rotates axes but Reshape only rearranges elements in memory. They are NOT interchangeable.

**TensorDot Operations**

**`tf.tensordot`** performs **tensor contraction**, which is a generalized form of the dot product and matrix multiplication.
It allows us to specify which **axes to contract over**, making it more flexible than **`tf.matmul`**.

**Tensor contraction means:**

- Multiply elements along shared axes  
- Then sum over those axes  
- Resulting tensor has reduced rank  

**It generalizes:**

- **Dot product** → Contract over **1 axis**
- **Matrix multiplication** → Contract over **last axis of A** and **second-last axis of B**
- **Higher-order tensor operations** → Contract over **multiple axes**


```python
# outer product
a = tf.constant([1, 2, 3])   # shape (3,)
b = tf.constant([4, 5])      # shape (2,)
outer = tf.tensordot(a, b, axes=0)
print(outer)
print("Shape:", outer.shape)
```

No contraction  

**Result shape → (3, 2)**  
Equivalent to **outer product**


```python
# basic dot product
a = tf.constant([1, 2, 3])   # shape (3,)
b = tf.constant([4, 5, 6])   # shape (3,)
result = tf.tensordot(a, b, axes=1)
print(result)
```

**Contract over 1 axis → standard dot product:**  
1 × 4 + 2 × 5 + 3 × 6 = **32**


```python
# matrix multiplication
A = tf.constant([[1, 2],
                 [3, 4],
                 [5, 6]])   # shape (3, 2)
B = tf.constant([[10, 20, 30],
                 [40, 50, 60]])  # shape (2, 3)
result = tf.tensordot(A, B, axes=[[1], [0]])
print(result)
```

**A axis 1** (size **2**)  
**B axis 0** (size **2**)  
**Result shape → (3, 3)** which is same as **`tf.matmul(A, B)`**

```python
# contracting over multiple axes
A = tf.random.uniform((2, 3, 4))  # shape (2, 3, 4)
B = tf.random.uniform((3, 4, 5))  # shape (3, 4, 5)
# Contract over axes (1,2) of A and (0,1) of B
result = tf.tensordot(A, B, axes=[[1, 2], [0, 1]])
print("Result shape:", result.shape)
```

**A axes:** (3, 4)  
**B axes:** (3, 4)  

Contracting removes those axes

**Remaining dims:**

- **A leftover:** (2,)
- **B leftover:** (5,)

**Result shape → (2, 5)**

This is impossible with **`tf.matmul`**, but we can do with **`tensordot`**.


#### Tensor Aggregations

```python
# get absolute vals
X = tf.constant([-9,-9, 7, 8.6])
print(f"abs val: {tf.abs(X)}")

# finx min max
X = tf.constant(np.random.randint(0, 100, 50))
print(f"min: {tf.reduce_min(X)}")
print(f"max: {tf.reduce_max(X)}")
print(f"mean: {tf.reduce_mean(X)}")
print(f"sum: {tf.reduce_sum(X)}")

print(f"variance: {tfp.stats.variance(X)}")

X = tf.cast(X, dtype=tf.float32)
print(f"std_dev: {tfp.stats.stddev(X)}") # this only works with real or complex nums

# or we can use tf.math instead of tfp module
print(tf.math.reduce_std(tf.cast(X, dtype=tf.float32)))
tf.math.reduce_variance(tf.cast(X, dtype=tf.float32))
```

in above operations since X is 1‑D, the reduction collapses the entire tensor to a scalar.

**What Does “Reduce” Mean?**

A **reduction operation**:

- Takes a tensor with **N dimensions**
- Aggregates values along a **specified axis**
- Returns a tensor with **fewer dimensions**

This is different from “normal” math function


```python
X = tf.constant([
    [1, 2, 3],
    [4, 5, 6]
])
print("X shape:", X.shape)

print(f"reduce over all elements: {tf.reduce_sum(X)}")
print(f"reduce over axis 0 (col-wise): {tf.reduce_sum(X, axis=0)}")
print(f"reduce over axis 1 (row-wise): {tf.reduce_sum(X, axis=1)}")
```

Axis 0
- **Column 1:** 1 + 4
- **Column 2:** 2 + 5  
- **Column 3:** 3 + 6

Axis 1
- **Row 1:** 1 + 2 + 3  
- **Row 2:** 4 + 5 + 6
